```python
import os
from netCDF4 import Dataset
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt

from tools.ColorPalette import make_palette_from_png, number_to_color, transform_and_map_color
```

```python
TEST_OUTPUT_DIR = "/work/pi_mzink_umass_edu/SPRITE/skillful_nowcasting/train/OUTPUT"
COLOR_PALLETE_PATH = "/work/pi_mzink_umass_edu/SPRITE/skillful_nowcasting/static/qpe_fine_rotated.png"
image_path = "/work/pi_mzink_umass_edu/SPRITE/skillful_nowcasting/netCDFData/train/20180908/20180908_135354.nc"

# Load the color palette
color_palette = make_palette_from_png(COLOR_PALLETE_PATH)
print(color_palette)


if not os.path.exists(TEST_OUTPUT_DIR):
    os.makedirs(TEST_OUTPUT_DIR)
```

```python
def save_frame_as_image(frame, x_data, y_data, filename, color_palette):
    vmin = 0  # Minimum value of your data range
    vmax = 203.2  # Adjust vmax based on the maximum value you're interested in

    # Transform and map each value in the frame to a color
    transformed_colored_frame = np.zeros((frame.shape[0], frame.shape[1], 4), dtype=np.uint8)
    for i in range(frame.shape[0]):
        for j in range(frame.shape[1]):
            transformed_colored_frame[i, j] = transform_and_map_color(frame[i, j], color_palette, vmin, vmax)

    # Optional: Display the image with matplotlib including colorbar for reference
    plt.figure(figsize=(10, 8))
    plt.imshow(
        transformed_colored_frame, extent=(x_data.min(), x_data.max(), y_data.min(), y_data.max()), origin="lower"
    )
    plt.colorbar(label="RRdata (mm/hr)")
    plt.title("NetCDF Data Plot")
    plt.xlabel("x0")
    plt.ylabel("y0")
    plt.savefig(os.path.join(TEST_OUTPUT_DIR, f"{filename}.png"))
    plt.close()


# Function to load a single frame from a NetCDF file and save it as an image
def load_and_save_frame(image_path, color_palette):
    with Dataset(image_path, "r") as nc_data:
        # Assuming 'RRdata', 'x0', and 'y0' are the variable names in your NetCDF file
        rr_data = nc_data.variables["RRdata"][:].squeeze()  # Remove the leading dimension if present
        x_data = nc_data.variables["x0"][:]
        y_data = nc_data.variables["y0"][:]

    filename = os.path.basename(image_path).split(".")[0]  # Remove file extension for the filename
    save_frame_as_image(rr_data, x_data, y_data, f"Original_{filename}", color_palette)


# Call the function with the specified image path and color palette
load_and_save_frame(image_path, color_palette)
```
